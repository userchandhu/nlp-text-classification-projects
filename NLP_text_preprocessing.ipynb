```python
# importing libraaries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
import warnings
warnings.filterwarnings('ignore')
```

# sample text

```python
# assigning text to a variable
text='my best friend name is prem and his worst friend name is chandu'
text
```

# tokenize

```python
# tolenization
words=word_tokenize(text)
words
```

# cleaning

```python
# text cleaning
cleaned_text=[word.lower() for word in words if word.isalpha()]
```

# stopwords removal

```python
# stopwords removal
stop_words=set(stopwords.words('english'))
filtered_words = [word for word in cleaned_text if word not in stop_words]
```

```python
# printing the outputs at each phase
print(text)
print('\n',words)
print('\n',filtered_words)
```
